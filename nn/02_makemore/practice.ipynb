- **Build a bigram**

```python
bigrCounts = {}
for w in words:
    chs = ['<S>'] + list(w) + ['<E>']
    for biagram in zip(chs, chs[1:]):
        biagramCount = bigrCounts.get(biagram, 0) + 1
        bigrCounts[biagram]["count"] = biagramCount

# sort by the counts
bigrams = sorted(bigrCounts.items(), key=lambda x: x[1], reverse=True)
bigrams[:10]
```

```python
# function to get bigram next char probabilities
def get_bigram_next_char_probs(bigrCounts, char, n):
    total_count = 0
    next_char_counts = {}
    
    for (first_char, second_char), count_dict in bigrCounts.items():
        if first_char == char:
            count = count_dict["count"]
            total_count += count
            next_char_counts[second_char] = next_char_counts.get(second_char, 0) + count
    
    # calculate probabilities
    next_char_probs = {char: count / total_count for char, count in next_char_counts.items()}
    
    # get top n probabilities
    sorted_probs = sorted(next_char_probs.items(), key=lambda x: x[1], reverse=True)[:n]
    
    return sorted_probs

bigram_next_char_probs = get_bigram_next_char_probs(bigrCounts, "e", 5)
bigram_next_char_probs
```
